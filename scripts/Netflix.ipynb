```python
import pandas as pd
```

```python
df = pd.read_csv('NetflixOriginals.csv')
df
```

```python
score = df.sort_values('IMDB Score', ascending=False)[:15]
score
```

```python
length = score[['Title','Runtime']]
length
```

```python
runtime = length.plot.bar(x='Title', y='Runtime',figsize=(20,10))
runtime.set_ylabel("minutes")
runtime.set_xlabel("title")
```

```python
sort_runtime = length.sort_values('Runtime', ascending=False)
sort_runtime
```

```python
sorted_runtime = sort_runtime.plot.bar(x='Title', y='Runtime',figsize=(20,10))
sorted_runtime.set_ylabel("minutes")
sorted_runtime.set_xlabel("title")
```

```python
sort_runtime = sort_runtime.set_index('Title')
```

```python
longest = sort_runtime.loc['Springsteen on Broadway']
longest
```

```python
shortest = sort_runtime.loc['Dancing with the Birds']
shortest
```

```python
difference = longest - shortest
difference
```

```python
difference[0]
```

```python
value = difference[0]
```

```python
print(f'The difference in runtime length between the longest and shortest movie is {value} minutes.')
```
